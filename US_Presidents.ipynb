# president_heights.csv

```python
#Import librerie (numpy, pandas) e del file .csv

import numpy as np
import pandas as pd

data = pd.read_csv(r'C:\Users\Marco Lombardini\Desktop\ComputerSociety\MaterialeEsercizi\president_heights.csv')
heights = np.array(data['height(cm)'])

print(heights)
```

```python
#Altezza massima, media, minima e Deviazione standard

print("Altezza media:", heights.mean())
print("Deviazione standard:", heights.std())
print("Altezza minima:", heights.min())
print("Altezza massima:", heights.max())
```

```python
#Primo, secondo e terzo quartile

print("Primo quartile:", np.percentile(heights, 25))
print("Media:", np.median(heights))
print("Terzo quartile:", np.percentile(heights, 75))
```

## Comandi per i grafici

```python
#Import delle librerie matplotlib.pyplot, seaborn, seaborn.set()

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set_theme()
```

```python
#Creazione del grafico con nome e rispettive coordinate

plt.hist(heights)
plt.title('Grafico di distribuzione Altezza Presidenti')
plt.xlabel('Altezza (cm)')
plt.ylabel('Numeri')

seaborn.set_style(style = "white")
seaborn.set_palette(palette = "PuBuGn")
```

```python
plt.hist(heights, bins= 20, histtype = "step",  orientation = "horizontal")
plt.show()
```

```python
seaborn.histplot(heights, binwidth= 5, stat="count", color="tab:red", kde=True, element= "step")
plt.show()
```
